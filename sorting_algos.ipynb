```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j+1], arr[j]
    return arr

def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        j = i - 1
        while j >= 0 and arr[j] > arr[j + 1]:
            arr[j + 1], arr[j] = arr[j], arr[j + 1]
            j = j - 1
    return arr
```

```python
def lcs(X, Y, m, n, lookup):

    # Initialise a key for the indices
    key = (m,n)
    if m == 0 or n == 0:
        # If either of the pointers reach  the starting of the string, then the LCS will have to be 0
        return 0

    # if the Key is not stored yet, then we estimate it. Otherwise it skips this part and returns the existing value
    if key not in lookup:

        # if the letters match, then the LCS will be 1 + the LCS corresponding to the two letters in front
        if X[m - 1] == Y[n - 1]:
            lookup[key] = lcs(X, Y, m -1, n - 1) + 1
        else:
            lookup[key] = max(lcs(X, Y, m - 1, n) , lcs(X, Y, m, n - 1)) 
        
    return lookup[key]
        
```

```python
def SCSLength(X, Y, m, n, lookup):
 
    # if the end of either sequence is reached, return
    # the length of another sequence
    if m == 0 or n == 0:
        return n + m
 
    # construct a unique key from dynamic elements of the input
    key = (m, n)
 
    # if the subproblem is seen for the first time, solve it and
    # store its result in a dictionary
    if key not in lookup:
 
        # if the last character of `X` and `Y` matches
        if X[m - 1] == Y[n - 1]:
            lookup[key] = SCSLength(X, Y, m - 1, n - 1, lookup) + 1
 
        # otherwise, if the last character of `X` and `Y` don't match
        else:
            lookup[key] = min(SCSLength(X, Y, m, n - 1, lookup),
                            SCSLength(X, Y, m - 1, n, lookup)) + 1
 
    # return the subproblem solution from the dictionary
    return lookup[key]
        
```

```python
def rod_length(prices : list, length = int):
    # Split the rods into two parts 
    n = len(prices)
    profit = [-1] * length
    max_profit = -10
    for i in range(1, length + 1):

        if profit[i] < 0 :
            profit[i] = max(max_profit, prices[i - 1] + rod_length(prices, n - i))
    
    return max_profit
```

```python
def coin_example(coins : list, sum_req : int):
    min_denomination = min(coins)
    if sum_req == min_denomination:
        return 1
    min_holder = 1e8
    for i in coins:
        if sum_req > i:
            chk = coin_example(coins, sum_req - i) + 1
            if chk < min_holder:
                min_holder = chk

    return chk
```

```python
def count(coins, N, sum):
# code here 
    if N <= 0:
        return 0
    
    if len(coins) == 0:
        return 0

    if sum < 0:
        return 0
    
    if sum == 0:
        return 1
    
    return count(coins, N - 1, sum) + count(coins, N, sum - coins[N-1])

count([1,2], 2, 3)
```

```python
def cutRod(price, n):
    #code here
    # initiliazing a lookup array
    t = [0 for i in range(n + 1)]
    
    # the maximum value for a rod of length 1 is its price
    t[1] = price[0]
    # We split the rod of length k into two parts - i and k - i where i can go from 0 to k - 1
    # We store the maximum value that can be arrived by splitting a rod of length k as t[k]
    # Now, any other rod of length more than k can be split into two rods - one that contains the price of i and the other that contains the max val of K - i
    for i in range(2,n + 1):
        for l in range(i):
            val = t[l] + price[i - l - 1]
            if val > t[i]:
                t[i] = val
        
        
    return t[n]
```

```python
def kadane(arr):
    current_sum, max_sum = 0, 0
    for i in range(len(arr)):
        if current_sum + arr[i] > 0:
            current_sum = current_sum + arr[i]
        else:
            current_sum = 0
        
        if current_sum > max_sum:
            max_sum = current_sum

    return max_sum
```

```python
def rearrange(arr, n): 
    ##Your code here
    s, e = 0, n - 1
    lst = []
    while e >= s:
        lst.append(arr[e])
        e = e - 1
        if e > s:
            lst.append(arr[s])
            s = s + 1
            
    return lst

rearrange([1,2,3,4,5,6,7], 7)
```

```python
def findMinDiff(A,N,M):

    # code here
    converted = list(set(A))
    print(converted)
    min = 1e8
    i = 0
    while i < N - M:
        if converted[i + M - 1] - converted[i] < min:
            min = converted[i + M - 1] - converted[i]
    
        i = i + 1
    
    return min

findMinDiff([3 ,4 ,1 ,9 ,56 ,7 ,9 ,12], 8, 5)
```

```python
def merge(arr1, arr2):

    m = len(arr1)
    n = len(arr2)

    lst = []

    i, j = 0, 0
    while (i < m  and j < n):
        if arr1[i] < arr2[j]:
            lst.append(arr1[i])
            i = i + 1
        else:
            lst.append(arr2[j])
            j = j + 1
    while i <= m - 1:
        lst.append(arr1[i])
        i = i + 1
    
    while j <= n - 1:
        lst.append(arr2[j])
        j = j + 1
    
    print(f'Unsorted : {arr1} & {arr2} --- Sorted : {lst}')
    return lst

def merge_sort(arr):

    n = len(arr)

    if n == 1:
        return arr
    
    mid = n//2

    left_sorted = merge_sort(arr[:mid])
    right_sorted = merge_sort(arr[mid:])

    return merge(left_sorted, right_sorted)

unsorted_list = [2, 4, 1, 5, 7, 2, 6, 1, 1, 6, 4, 10, 33, 5, 7, 23]
sorted_list = merge_sort(unsorted_list)
print(unsorted_list)
print(sorted_list)
```

```python
#Function is to check whether two strings are anagram of each other or not.
def isAnagram(a,b):
    #code here
    
    dict1 = dict()
    dict2 = dict()
    
    if len(a) != len(b):
        return 'NO'
    
    for i in range(len(a)):
        
        if a[i] in dict1.keys():
            dict1[a[i]] = dict1[a[i]] + 1
        else:
            dict1[a[i]] = 1

        if b[i] in dict2.keys():
            dict2[b[i]] = dict2[b[i]] + 1
        else:
            dict2[b[i]] = 1
            
    
    print(dict1, dict2)
    if dict1 == dict2:
        return 'YES'
    else:
        return 'NO'

isAnagram('moron', 'ronmo')
```

```python
def binary_search(arr, num):

    n = len(arr)
    first = 0
    last = n - 1
    while (last >= first):
        mid = (first + last)//2
        print(first,mid,last)
        if arr[mid] == num:
            return mid
        elif arr[mid] > num:
            last = mid - 1
        else:
            first = mid + 1
    
    return -1

binary_search([1,5,6,8,16,22,24,27,29,31,35], 35)
```

```python
def jump_search(arr, num):

    n = len(arr)
    jump = int(n ** (0.5))
    l, r = 0, jump
    if num > arr[n - 1] or num < arr[0]:
        return -1

    print('Initial pointers', l, r)
    while l <= n - 1:
        r = min(r, n - 1)
        if arr[l] <= num <= arr[r]:
            print('Checking between', l, r + 1)
            for i in range(l,r + 1):
                if arr[i] == num:
                    return i
            return -1
        else:
            l = l + jump
            r = r + jump
            print('Incrementing jump', l, r)

jump_search([1,5,6,8,16,22,24,27,29,31,35], 35)
```
